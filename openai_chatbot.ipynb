```python
COMPLETIONS_MODEL = "text-davinci-003"
EMBEDDINGS_MODEL = "text-embedding-ada-002"
openai.api_key = getpass("API Key") 
```

```python
df1 = pd.read_csv('Finance_data .csv')
df1.head()
```

```python
from openai.embeddings_utils import get_embedding

df1['vectors'] = df1['Questions'].apply(lambda row:get_embedding(row,engine=EMBEDDINGS_MODEL))
df1.head()
```

```python
question ='explain me about debt-equity ratio?'
vector = get_embedding(question,engine=EMBEDDINGS_MODEL)
```

```python
vector
```

```python
from openai.embeddings_utils import cosine_similarity

df1['cosine_similarity'] = df1['vectors'].apply(lambda x: cosine_similarity(x,vector))
df1.head()
```

```python
answer = df1.sort_values('cosine_similarity',ascending=False).head(5)
answer
```

```python
answers = []
for i,row in answer.iterrows():
  answers.append(row['Answers'])
```

```python
answers
```

```python
model_engine = COMPLETIONS_MODEL
model_prompt = f"only using the context given here Answer the question in Warren buffet style context :{answers}\n\nQuestion: {question}\nAnswer:"

response = openai.Completion.create(
  engine=model_engine,
  prompt=model_prompt,
  max_tokens=100,
  n=1,
  stop=None,
  temperature=0.5,
)

answer = response.choices[0].text.strip()
```
